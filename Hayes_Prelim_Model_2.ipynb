# What is HIV and How Does One Become Infected?

The HIV virus is a bloodborne pathogen, meaning that one can only contract the virus if he or she comes into contact with an infected individual's blood. This could occur through sharing needles, through the birthing process from mother to baby, or from unprotected sex with someone who is infected.

---


HIV is an in-host viral infection, meaning that it cannot survive outside of its host's body. Once ouside of the host's body , approximately nintey to ninety-nine percent of the virus will become inactive. This means that any contact with the virus outside of infected blood is not sufficient enough to infect an individual.

### What Does HIV Do to the Body?

Once inside its host, the Human Immunodeficiency Virus targets specific white blood cells known as T cells, more specifically CD4 cells, that play a vital role in activating or suppressing one's immune system. The HIV virions are replicated within these CD4 cells. Thus, the CD4 cells are ineffective and inevitably destroyed. As the CD4 cell count diminishes, the body's immune system weakens and is unable to fight off infection. Therefore, the progression of HIV to AIDs takes approximately ten years.

# Modeling Viral Infections Such as HIV:

### How Does a Network Model Work?

Viral compartmental models often work by separating the population within a certain network of individuals into specific categories based on the differential equations used to model the movement between specific populations within that network. These equations are fit to effectively match the real-world that these populations of individuals are based off of. The rate of transition between categories of individuals is based off of environmental parameters that exist within the real world network the system is meant to model. For example, if there is a virus that exists which statistically infects 50% of individuals who interact with an infected individual, there would exists an infectivity parameter within the model with the value of 0.5 or 50%.

If we wanted to model a very basic differential equation set which simply shows the rate of change from Susceptible individuals to Infected individuals (S -> I), we could simply write:

###### dS/dt = −βSI 

###### dI/dt = βSI 

Where the parameter, β refers to the rate of contact between Infected and Susceptible individuals within the model. These equations allow us to calulate the rates of change of these populations (I and S) over time. These equations also depend on the initial quantities of each population within the model, as increasing the initial number S individuals or I individuals in this example would cause an increase in the rates of change of the populations. This type of transition is often more easily described by a diagram showing the shift from one population to another based on a certain equation. For example:

Models such as these can range from very simple (SI, SIR) models to very complex. The versatile complexity allows for the introduction of concepts such as vaccines, quarantines, viral death rate, etc. into a network model. This is how these models are able to effectively represent real world scenarios that may exist, and can often help to explain patterns in certain model behaviors that can allow us to more effectively predict and analyze real world events such as viral outbreaks.

Of course, mathematical models often struggle to take into account certain aspects of the real world that may effect whatever the model is attempting to replicate. Because of this, mathematical models are not perfect, but they are often as close as we can come to accurately predicting many real world networks and events.

### Basic Reproduction Number (R0)

A major factor in viral models is the basic reproduction number, or R0. R0 is the number of individuals a given infected person will infected during the period of infectibility of whatever virus they carry. The larger the R0 number, the faster the virus will spread and visa versa. If R0 < 1, the viral spread is decreasing, whereas if R0 > 1, the spread is increasing. This number is often used to predict wether the spread of a real-world infection will increase or decrease based on certain environmental parameters put into place. R0 can generally be calculated as:

###### R0 = β/μ

in which β is the rate of contact between individuals and μ is the recovery rate of an infected individual. However, the calculations for R0 can become much more complicated as more modes of infection and recovery become introduced to a network, but this general calculation is often all that is required.

```python
import tellurium as te
import ipywidgets as widgets
from IPython.display import display
HIV_model = '''
  //Equations
  E1: T -> I ; beta * V * T ; //infection rate
  E2: T -> ; rho * T ; //death rate of health T-Cells
  E3: -> T ; s + alpha * T * (1 - (T + I)/ Tmax) ; //supply of healthy cells 
  E4: I -> ; I * chi ; //death rate of infeced T-cells
  E5: V -> ; V * tau ; //death rate of virions
  E6: -> V ; gamma1 * chi * I ; //replication rate of HIV virus
    
  //Parameters
  s = 0.1 ; //new supply of healthy cells, from thymus
  rho = 0.02 ; //death rate of healthy cells
  alpha = 3 ; //replication rate of healthy T-cells
  Tmax = 1500 ; //maximum level of healthy cells
  beta = 0.0027 ; //infection rate
  chi = 0.3 ; //death rate of infected cells
  gamma1 = 10 ; //replication rate of virus
  tau = 2.4 ; //death rate of free virions
    
  //Initial Conditions
  T = 0.1 ;
  I = 0 ; 
  V = 0.1 ; 
'''
    
m = te.loada(HIV_model)
s = m.simulate(0, 70, 7000)
    
m.plot()
```

```python
print("S0, Thymus")
# Range of the parameter sweep, need to specify lowest value, highest value and number of points 
import numpy as np

lowest_value = 0.1
highest_value = 0.5
num_points = 100
B = np.linspace(lowest_value,highest_value,num_points)
# Placeholder matrix to record output metric, first column is parameter value, second column is output metric
R = np.zeros((len(B),2))
# Iterate over the range of the parameter value
for i in range(len(B)):
    # Load model
    m = te.loada(HIV_model)
    # Assign value to the parameter in the model
    m.s = B[i]
    # Simulate model
    s = m.simulate(0,100,1000,['time','T','I','V'])
    # Record value of the parameter in the model
    R[i,0] = m.s
    # Record output metric (e.g. fraction of the population that got infected)
    R[i,1] = s['I'][-1]
# Plot results of parameter sweep
te.plotArray(R, title='Parameter Sweep, s0',grid='True',ylabel = 'Final Number of Infected Cells',xlabel = 'Number of Healthy T-Cells Released by Thymus/Day')
```

```python
def model_slider(s, alpha, gamma1):
    HIV_model = '''
        //Equations
        E1: T -> I ; beta * V * T ; //infection rate
        E2: T -> ; rho * T ; //death rate of health T-Cells
        E3: -> T ; s + alpha * T * (1 - (T + I)/ Tmax) ; //supply of healthy cells 
        E4: I -> ; I * chi ; //death rate of infeced T-cells
        E5: V -> ; V * tau ; //death rate of virions
        E6: -> V ; gamma1 * chi * I ; //replication rate of HIV virus
    
        //Parameters
        s = 0.1 ; //new supply of healthy cells, from thymus
        rho = 0.02 ; //death rate of healthy cells
        alpha = 3 ; //replication rate of healthy T-cells
        Tmax = 1500 ; //maximum level of healthy cells
        beta = 0.0027 ; //infection rate
        chi = 0.3 ; //death rate of infected cells
        gamma1 = 10 ; //replication rate of virus
        tau = 2.4 ; //death rate of free virions
    
        //Initial Conditions
        T = 0.1 ;
        I = 0 ; 
        V = 0.1 ; 
     '''
    
    m = te.loada(HIV_model)
    m['s'] = s
    m['alpha'] = alpha
    m['gamma1'] = gamma1
    s = m.simulate(0, 70, 7000)
    
    m.plot()
```

```python
#Altering Parameters in Time (slider definitions)
s_slider = widgets.FloatSlider(
                            min = 0.0, 
                            max = 2, 
                            step = 0.01, 
                            description='s',
                            value=m['s'])
alpha_slider = widgets.FloatSlider(
                            min = 0.0, 
                            max = 10, 
                            step = 0.1, 
                            description='alpha',
                            value=m['alpha'])
gamma1_slider = widgets.FloatSlider(
                            min = 0.0, 
                            max = 20, 
                            step = 0.2, 
                            description='gamma1',
                            value=m['gamma1'])
```

```python
r = widgets.interact_manual(model_slider, s=s_slider, alpha=alpha_slider, gamma1=gamma1_slider)
```
